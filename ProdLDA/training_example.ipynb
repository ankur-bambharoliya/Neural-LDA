```python
from train2 import train, print_perp
import torch
from hyper_parameters import HyperParameters, USE_CUDA
from prod_lda import ProdLDA
hps = HyperParameters()
```

```python

model = ProdLDA(hps)

if USE_CUDA:
    model.cuda()

if hps.optimizer == 'Adam':
    optimizer = torch.optim.Adam(model.parameters(), hps.learning_rate,
                                 betas=(hps.momentum, 0.999))
else:
    optimizer = torch.optim.SGD(model.parameters(),
                                hps.learning_rate, momentum=hps.momentum)

train(model, optimizer)
print_perp(model)
```
